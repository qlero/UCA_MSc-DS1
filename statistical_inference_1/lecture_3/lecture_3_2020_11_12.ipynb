##  Context

# REWRITE1

$$P(\bar{X_n}-\mu\in[-T_n,S_n])=1-\alpha \quad\Leftrightarrow\quad P(\sqrt{n}.\frac{\bar{X_n}-\mu}{\hat{\sigma_n}}\in[-\frac{\sqrt{n}.T_n}{\hat{\sigma_n}},\frac{\sqrt{n}.S_n}{\hat{\sigma_n}}])=1-\alpha$$
and:
$$\sqrt{n}.\frac{\bar{X_n}-\mu}{\hat{\sigma_n}}\sim ð’¯(n-1)$$
with:
$$\alpha_1 = P(\sqrt{n}.\frac{\bar{X_n}-\mu}{\hat{\sigma_n}} \lt -\frac{\sqrt{n}.T_n}{\hat{\sigma_n}})$$
$$\alpha_2 = P(\sqrt{n}.\frac{\bar{X_n}-\mu}{\hat{\sigma_n}} \gt \frac{\sqrt{n}.S_n}{\hat{\sigma_n}})$$

# REWRITE2

$$[\bar{X_n}-\frac{\hat{\sigma_n}}{\sqrt{n}}.t_{1-\alpha_1}\quad,\quad\bar{X_n}+\frac{\hat{\sigma_n}}{\sqrt{n}}.t_{1-\alpha_2}]$$
and:
$$t_{1-\alpha_1}=t_{1-\alpha_1, n-1}\quad t_{1-\alpha_2}=t_{1-\alpha_2, n-1}$$

$$P(\bar{X_n}-\mu\in[-T_n,S_n])=1-\alpha \quad\Leftrightarrow\quad P(\sqrt{n}.\frac{\bar{X_n}-\mu}{\hat{\sigma_n}}\in[-\frac{\sqrt{n}.T_n}{\hat{\sigma_n}},\frac{\sqrt{n}.S_n}{\hat{\sigma_n}}])=1-\alpha$$

$$P(ð“(0,1)\in[-\frac{\sqrt{n}.T_n}{\hat{\sigma_n}},\frac{\sqrt{n}.S_n}{\hat{\sigma_n}}])=1-\alpha$$

# Simple Linear Regression

##  Context

$\forall i\in[1,n]$ observations $(x_i, y_i)$: 

- $x_i$ is an observation of $X_i$ 
- $y_i$ is an observation of $Y_i$. 

We assume $(X_1, Y_1), ..., (X_n, Y_n)$ are **IID random variables**.

We denote by $(X,Y)$ the common distribution of the $(X_i, Y_i)$ pair. In practice **we do not know the distributions of $X$, $Y$, or $(X,Y)$**. Thus the scope of simple linear regression is:

> The identification of a possible link between $X$ and $Y$

## Linear Model

<u>The scope is:</u>

> We want to find the best function that can be written as $f(x)=a.x+b$ such that $Y \approx f(x)$

<u>To do this:</u>

The considered model is: $Y=a.X+b+\epsilon$ with:

- $\epsilon$ (noise) a RV
- $Y$ is [consequently] a RV
- $X$ can be a random variable but <span style="color:red">**in the following, we take X as a fixed design; we do not consider $X$ as a RV**</span>.
- $a$ and $b$ are unknown and must be estimated

<u>Assumptions on the model:</u>

- $\mathbb{E}[\epsilon_i]=0$ for $i \in [1, n]$

- $\mathbb{V}[\epsilon_i]=\sigma^2$ for $i \in [1, n]$

- $cov(\epsilon_i, \epsilon_j)=0$ if $i\neq j$

where:

$$Y_i=a.X_i+b+\epsilon_i$$

At some moment, we will consider an additional assumption: 

- $\epsilon_i \sim N(0,\sigma^2)$, knowing we have no observation of $\epsilon_i$

### How to estimate $a$ and $b$?

We can use the **least square criterion**. 

**We want to determine $a_n$ and $b_n$ such that if we denote $\hat{y_i} = \hat{a_i}.X_i + \hat{b_i}$ then all the $\hat{y_i}$ are going to be close to $Y_i$**.

```python
Image("pictures/ex_lin_reg_1.png", width=600)
```

We consider the **least square criterion**:

$$S(a,b) = \underset{i=1}{\overset{n}{\sum}}(y_i - (a.X_i+b))^2$$

To minimize the error between each observation $y_i$ and the model prediction $\hat{y_i}$, We define $\hat{a_n}$ and $\hat{b_n}$ as:

> $S(\hat{a_n}, \hat{b_n}) = \underset{a,b}{min} S(a,b) \rightarrow$ optimization problem

**Theorem**:

The solution of the least square problem is:

- <span style="color:orange">$\hat{b_n}=Y\bar{Y_n}-\hat{a_n}.\bar{X_n}$</span>

$$\hat{a_n} = \frac{\underset{i=1}{\overset{n}{\sum}}X_i.Y_i - n.\bar{X_n}\bar{Y_n}}{\underset{i=1}{\overset{n}{\sum}}(X_i-\bar{X_n})^2} = \frac{\underset{i=1}{\overset{n}{\sum}}X_i.Y_i - n.\bar{X_n}\bar{Y_n}}{\underset{i=1}{\overset{n}{\sum}}X_i^2-n.\bar{X_n}^2} =\frac{\underset{i=1}{\overset{n}{\sum}}(Y_i-\bar{Y_n})(X_i-\bar{X_n})}{\underset{i=1}{\overset{n}{\sum}}X_i^2-n.\bar{X_n}^2}$$

**Properties**: 

- $\mathbb{E}[\hat{a_n}] = a$
- $\mathbb{E}[\hat{b_n}] = b$
- $\mathbb{V}[\hat{a_n}] = \sigma_b^2$
- $\mathbb{Cov}[\hat{a_n}, \hat{b_n}] = \sigma_{a,b}^2$

$$
\sum=\begin{bmatrix} 
\sigma_b^2 & \sigma_{a,b}^2 \\
\sigma_{a,b}^2 & \sigma_a^2 \\
\end{bmatrix} = \sigma^2(^t\mathbb{X}\mathbb{X})^{-1}
\quad
$$ with $$\mathbb{X}=\begin{pmatrix} 
1 & x_1 \\
1 & x_2 \\
... & ... \\
1 & x_n
\end{pmatrix}$$ 2 columns and n rows.

**Remark**: 

$$Cov(\hat{a_n},\hat{b_n})=\mathbb{E}[(\hat{a_n}-\mathbb{E}[\hat{a_n}]).(\hat{b_n}-\mathbb{E}[\hat{b_n}])]$$

$$Cov(\hat{a_n},\hat{a_n})=\mathbb{V}[\hat{a_n}]$$

If $\hat{a_n}$ and $\hat{b_n}$ are independent then $Cov(\hat{a_n},\hat{b_n})=0$

**Remark**: 

The model is $\forall i \in [1,n], y_i=a.X_i+b+\epsilon_i$. Written with matrices:

$$\mathbb{Y}=\begin{pmatrix}y_1 \\... \\ y_n\end{pmatrix}\quad\mathbb{U}=\begin{pmatrix}\epsilon_1 \\... \\ \epsilon_n\end{pmatrix}\quad\beta=\begin{pmatrix}b \\ a\end{pmatrix}$$

we have:

$\mathbb{Y} = \mathbb{X}.\beta + \mathbb{U}$ matricial writing of a linear model

$$S(\beta)=\underset{i=1}{\overset{n}{\sum}}(y_i-(a.X_i+b))^2=(y_1-(a.X_1+b), ..., y_n-(a.X_n+b)).\begin{pmatrix} 
y_1-(a.X_1+b) \\... \\ y_n-(a.X_n+b)\end{pmatrix}=\mathbb{Y}-\mathbb{X}\beta$$

We have:

$$\mathbb{Y}=\mathbb{X}\beta+\mathbb{U}$$

Thus: 

$$S(\beta) \quad=\quad ^t(\mathbb{Y}-\mathbb{X}.\beta) * (\mathbb{Y}-\mathbb{X}.\beta) \quad=\quad ^t\mathbb{Y}.\mathbb{Y} - ^t\mathbb{Y}.\mathbb{X}.\beta - ^t\beta.^t\mathbb{X}.\mathbb{Y} + ^t\beta.^t\mathbb{X}.\mathbb{X}.\beta$$

and: 

- $^t\mathbb{Y}.\mathbb{Y}$ does not depend on $\beta \Rightarrow$ we can forget this term when we try to find the best $\beta$
- $^t\mathbb{Y}.\mathbb{X}.\beta = ^t\beta.^t\mathbb{X}.\mathbb{Y}$

we want to minimize $S(\beta)$ with respect to $\beta$

i.e. We want to determine $\beta$ that minimize:

$$\tilde{S}(\beta) = ^t.\beta^t.\mathbb{X}.\mathbb{X}.\beta-2.^t\mathbb{Y}.\mathbb{X}.\beta$$

$$\nabla\tilde{S}(\beta)=2.^t.\mathbb{X}.\mathbb{X}.\beta-2.^t.\mathbb{X}.\mathbb{Y}$$

We have to find: $$\nabla\tilde{S}(\beta)=0$$

The solution denoted $\hat{\beta_n}$ is:

$$\hat{\beta_n}=(^t.\mathbb{X}.\mathbb{X})^{-1}.^t\mathbb{X}.\mathbb{Y}$$

<span style="color:red">/!\\</span> $^t.\mathbb{X}.\mathbb{X}$ should be invertible:

$$\mathbb{X} = \begin{pmatrix} 
1 & x_1 \\
1 & x_2 \\
... & ... \\
1 & x_n
\end{pmatrix}$$

$^t\mathbb{X}.\mathbb{X}$ will be invertible if there exists $i \neq x_k$

### Proof of $\mathbb{E}[\hat{\beta_n}]=\beta$ and $\mathbb{V}[\hat{\beta_n}]=\sigma^2.(^t\mathbb{X}.\mathbb{X})^{-1}$

Given $\hat{\beta_n}$ a random vector, we have:

- $\mathbb{E}[\hat{\beta_n}] = \begin{pmatrix} \mathbb{E}[\hat{b_n}] \\ \mathbb{E}[\hat{a_n}] \end{pmatrix}$
- $\hat{\beta_n} = \begin{pmatrix} \hat{b_n} \\ \hat{a_n} \end{pmatrix}$

$\mathbb{E}[\hat{\beta_n}]=\mathbb{E}[(^t\mathbb{X}.\mathbb{X})^{-1}.^t\mathbb{X}.(\mathbb{X}.\beta+\mathbb{U})]$

$=\mathbb{E}[(^t\mathbb{X}.\mathbb{X})^{-1}.(^t\mathbb{X}.\mathbb{X}).\beta+(^t\mathbb{X}.\mathbb{X})^{-1}.^t\mathbb{X}.\mathbb{U}]$

$=\mathbb{E}[\beta+(^t\mathbb{X}.\mathbb{X})^{-1}.^t\mathbb{X}.\mathbb{U}]$

$=\mathbb{E}[\beta]+\mathbb{E}[(^t\mathbb{X}.\mathbb{X})^{-1}.^t\mathbb{X}.\mathbb{U}]$

$=\beta+(^t\mathbb{X}.\mathbb{X})^{-1}.^t\mathbb{X}.\mathbb{E}[\mathbb{U}]$

$=\beta+(^t\mathbb{X}.\mathbb{X})^{-1}.^t\mathbb{X}.0_n$

Thus:

$\mathbb{E}[\hat{\beta_n}]=\beta$

**Remember**:

- $T$ a RV
- $A$ a fixed matrix
- $B$ a fixed matrix

$\mathbb{E}[\mathbb{A}.\mathbb{T}] = \mathbb{A}.\mathbb{E}[\mathbb{T}]$

$\mathbb{E}[\mathbb{B}.\mathbb{T}] = \mathbb{B}.\mathbb{E}[\mathbb{T}]$

$\mathbb{M}$ a random matrix. $\mathbb{A}$ and $\mathbb{B}$ two fixed matrices. Then:

$\mathbb{E}[\mathbb{A}.\mathbb{M}.\mathbb{B}] = \mathbb{A}.\mathbb{E}[\mathbb{M}].\mathbb{B}$


$$\mathbb{E}[\mathbb{T}]=\begin{pmatrix} \mathbb{E}[\mathbb{T_1}] \\ ... \\ \mathbb{E}[\mathbb{T_d}] \end{pmatrix}$$ 

if:

$$T=\begin{pmatrix} \mathbb{T_1} \\ ... \\ \mathbb{T_d} \end{pmatrix}$$

$$\mathbb{M}=(\mathbb{M}_{ij}) \Rightarrow \mathbb{E}[\mathbb{M}] = (\mathbb{M_{ij}})$$

$\mathbb{V}[\hat{\beta_n}] = \mathbb{E}[(\hat{\beta_n}-\mathbb{E}[\hat{\beta_n}]).^t(\hat{\beta_n}-\mathbb{E}[\hat{\beta_n}])]\rightarrow$ we obtain a matrix.
