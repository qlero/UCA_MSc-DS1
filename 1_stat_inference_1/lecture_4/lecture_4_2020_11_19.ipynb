# Linear Model

$Y=\beta_0 + \beta_1.X+\epsilon$ with noise $\epsilon$ and $\forall i \in [1, n], \mathbb{E}[\epsilon_i]=0$, $\mathbb{V}[\epsilon_i]=\sigma^2$ and $cov(\epsilon_i, \epsilon_j)=0$

### Matricial writing

We recall the matricial writing is written as such:

$Y = \mathbb{X}\beta+U$ with $\mathbb{Y}=\begin{pmatrix}y_1 \\... \\ y_n\end{pmatrix}$, $\mathbb{U}=\begin{pmatrix}\epsilon_1 \\... \\ \epsilon_n\end{pmatrix}$, $\beta=\begin{pmatrix}\beta_0 \\ \beta_1\end{pmatrix}$ and $\mathbb{X} = \begin{pmatrix} 
1 & x_1 \\
1 & x_2 \\
... & ... \\
1 & x_n
\end{pmatrix}$
We have the least square solution:

$$\hat{\beta_n}=(^t.\mathbb{X}.\mathbb{X})^{-1}.\mathbb{X}.\mathbb{Y}$$

**Remark 1**: $(^t.\mathbb{X}.\mathbb{X})^{-1} \neq .\mathbb{X}^{-1}.(^t\mathbb{X})^{-1}$ because $\mathbb{X}^{-1}$ generally does not exist.

$$\hat{\beta_n}=\begin{pmatrix}\hat{\beta}_0 \\ \hat{\beta}_1\end{pmatrix}$$

$$\mathbb{V}(\hat{\beta}_n = \begin{pmatrix} 
\mathbb{V}(\hat{\beta}_0) & Cov(\hat{\beta}_0, \hat{\beta}_1) \\
Cov(\hat{\beta}_0, \hat{\beta}_1) & \mathbb{V}(\hat{\beta}_1)
\end{pmatrix}$$

$$\mathbb{V}(\hat{\beta_0})=\frac{\sigma^2.\underset{i=1}{\overset{n}{\sum}}X_i^2}{n.\underset{i=1}{\overset{n}{\sum}}(X_i-\bar{X_n})^2}$$

$$\mathbb{V}(\hat{\beta_1})=\frac{\sigma^2}{\underset{i=1}{\overset{n}{\sum}}(X_i-\bar{X_n})^2}$$

$$Cov(\hat{\beta}_0, \hat{\beta}_1)=\frac{\sigma^2.\bar{X}_n}{\underset{i=1}{\overset{n}{\sum}}(X_i-\bar{X_n})^2}$$

If the value of $\sigma^2$ is small, the values of $\hat{\beta_0}$ and $\hat{\beta_1}$ become small as well. It implies the observations X_i should be large to reduce the size of $V[\hat{\beta_n}]$. Those variances are theoretical quantities since they depend on $\sigma^2$ which is unknown.

As such **we need to estimate $\sigma^2$**. 

### Estimating $\sigma^2$

$Y_i=\beta_0 + \beta_1.X_i+\epsilon \Rightarrow \mathbb{E}[Y_i] = \beta_0 + \beta_1.X_i$ knowing that the $Y_i$ are not identically distributed. However $\forall i\in [1,n] \rightarrow \mathbb{V}[Y_i] = \sigma^2$.

If we consider $\mathcal{Z}_i = Y_i -(\beta_0 + \beta_1.X_i)$ with:
- $\mathbb{E}[\mathcal{Z}_i] = 0$
- $\mathbb{V}[\mathcal{Z}_i] = \sigma^2$

But we do not have access to those $\mathcal{Z}_i$. So we're going to introduce the **residuals**.

**Definition**:

$$\epsilon_i = Y_i - \hat{Y_i}$$ with: $$\hat{Y_i}=\hat{\beta_0} + \hat{\beta_1}.X_i$$ with: $$\hat{Y_i},\,the\,prediction\,of\,Y\,for\,individual\,i$$

- **Property**: $\mathbb{E}[\epsilon_i] = 0$

- **Proposition**: An unbiased estimator of $\sigma^2$ is given by: $\hat{\sigma^2_n} = \frac{1}{n-2}\underset{i=1}{\overset{n}{\sum}}\hat{\epsilon}_i^2$

### Prevision

**Main aim of linear model**: if we have a new observation for $X_1$, a new observation denoted $x_{n+1}$ we would like to get the associated value for $Y$, denoted $y_{n+1}$, knowing we don't have access to the true value of $y_{n+1}$.

As such we assume that we have $y_{n+1}=\beta_0+\beta_1.x_{n+1} + \epsilon_{n+1}$ and $\mathbb{E}[\epsilon_{n+1}] = 0$, $\mathbb{V}[\epsilon_{n+1}] = \sigma^2$ and $\forall i \in [1, n], Cov[\epsilon_{n+1}, \epsilon_{i}]=0$

One way to predict $y_{n+1}$ is to use $\hat{y^p_{n+1}} = \hat{\beta_0}+\hat{\beta_1}.x_{n+1}$

The $^p$ is to say that this is a prevision since $x_{n+1}$ has not been used in the construction of $\hat{\beta_0}$ and $\hat{\beta_1}$.

**Proposition**:

$$\mathbb{V}[\hat{y}^p_{n+1}] = \sigma^2.\frac{1}{n}.\frac{(x_{n+1}-\bar{x_n})^2}{\underset{i=1}{\overset{n}{\sum}}(x_{n+1}-\bar{x_n})^2}$$

The **prevision error** is: 
$$\hat{\epsilon}^p_{n+1} = y_{n+1}-\hat{y^p_{n+1}}$$ With:
$$\mathbb{E}[\hat{\epsilon}^p_{n+1}]=0$$ And:
$$\mathbb{V}[\hat{\epsilon}^p_{n+1}] = \sigma^2.(1+\frac{1}{n}+\frac{(x_{n+1}-\bar{x_n})^2}{\underset{i=1}{\overset{n}{\sum}}(x_{n+1}-\bar{x_n})^2})$$


**Remark 3**: $\mathbb{V}[\hat{\epsilon}^p_{n+1}] = \sigma^2+\mathbb{V}[\hat{y}^p_{n+1}]$
The reason is that $\hat{y}^p_{n+1} = \hat{\beta_0} + \hat{\beta_1}.X_{n+1}$. There is one error of estimation but only when we consdier that the model is the reality.

The error if the model is the reality is:

$$\hat{\epsilon}^p_{n+1} = y_{n+1} - \hat{y}^p_{n+1} = \beta_0 + \beta_1.X_{n+1}+\epsilon_{n+1}-(\hat{\beta}_0+\hat{\beta}_1.X_{n+1})$$ 

with $\epsilon_{n+1}$ the quantificaition of thegap between the reality and the model.

We just have one more thing to see with just the minimal ssumptions on the noise. This is: **How to quantify the quality of the linear model**?

### Geometrical interpretation of the linear model

Representation of the variables:

- $\mathbb{R}^n$: the space of the variables
- $\mathcal{E}_{\mathbb{X}}$: the linear subspace generated by the columns of $\mathbb{X}$

In the simple linear model $\mathcal{E}_{\mathbb{X}}$ is generated by $\mathbb{1}=\begin{pmatrix} 1 \\ ... \\ 1 \end{pmatrix}$ and the vector $\mathbb{X}=\begin{pmatrix} X_1 \\ ... \\ X_n \end{pmatrix}$ and: 

- $dim(\mathcal{E}_{\mathbb{X}})=2$

```python
Image('pictures/representation.PNG', width=600)
```

Because of the **least square criterion**, $\hat{\mathbb{Y}}$ is the point of $\mathcal{E}_{\mathbb{X}}$ which is the closest to $\mathbb{Y}$.

$\Rightarrow$ $\hat{\mathbb{Y}}$ is the project of $\mathbb{Y}$ on $\mathcal{E}_{\mathbb{X}}$

**Remark 4**: The projection of $\mathbb{Y}$ on the linear subspace generated by $\begin{pmatrix} 1 \\ ... \\ 1 \end{pmatrix}$ is $\bar{y}_n.\begin{pmatrix} 1 \\ ... \\ 1 \end{pmatrix}$
the projection of $\hat{Y}$ on the same linear subspace is also $\bar{y}_n.\begin{pmatrix} 1 \\ ... \\ 1 \end{pmatrix}$

With the triangle ABC in the space C. Thanks to the Pythagorean theorem we have:

the total variation in an observed response about its mean can be written as a sum of two parts - its deviation from the fitted value plus the deviation of the fitted value from the mean response $y_i \bar{y} = (y_i  \hat{y}_i) + (\hat{y}_i  \bar{y})$

$$Total\,Var\,of\,Y = Residual\,Variance + Variance\,Explained\,by\,the\,Model$$
$$SST = SSR + SSE$$
$$\underset{i=1}{\overset{n}{\sum}}(y_i-\bar{y}_n)^2 = \underset{i=1}{\overset{n}{\sum}}(\hat{y}_i-\bar{y}_i)^2+\underset{i=1}{\overset{n}{\sum}}(y_i-\hat{y}_n)^2$$
$$\underset{i=1}{\overset{n}{\sum}}(y_i-\bar{y}_n)^2 = \underset{i=1}{\overset{n}{\sum}}\hat{\epsilon}_i^2+\underset{i=1}{\overset{n}{\sum}}(y_i-\hat{y}_n)^2$$

SSR should be small.

We introduce the **coefficient of determination $R^2$** with:

$$R^2=\frac{\underset{i=1}{\overset{n}{\sum}}(\hat{y}_i-\bar{y}_i)^2}{\underset{i=1}{\overset{n}{\sum}}(y_i-\bar{y}_n)^2}=\frac{SSE}{SST}$$

**Properties**:

- $R^2\in[0,1]$
- $R^2=cos^2\theta$
- if $R^2=1$ then $\theta=0\Rightarrow Y\in\mathcal{E}_{\mathbb{X}}$ (linear model is a reality)
- if $R^2=0$ then $\underset{i=1}{\overset{n}{\sum}}(\hat{y}_i-\bar{y}_i)^2=0\Rightarrow\hat{y}_i=\bar{y}_i$ (linear model is a totally wrong)
- if $R^2$ is near 0, it means that $Y$ is quite in $(\mathcal{E}_{\mathbb{X}})^\perp$ and the linear model is wrong
- if $R^2$ is near 1, it means that $Y$ is quite in $(\mathcal{E}_{\mathbb{X}})$ and the linear model is correct
