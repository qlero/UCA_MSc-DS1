## 5. Guidelines

### 5.1. Title, Authors, Acknowledgements

1. The Title should be very descriptive -- often stating the main finding(s) -- and is not about being creative or "catchy."
2. The order of authors is important

To avoid misleading statements of authors, provide the name of the people who provided both work and insights. 

Acknowledgements are for helping technicians, etc. for instance. It also includes who funded the research.

### 5.2. Abstract/Summary

It must contain:

1. Brief background of the subject
2. Purpose of the study
3. Major findings of the study
4. Relationship between these findings and the field

### 5.3. Introduction

It must present the background information for a fellow scientist (possibly an expert in another field alltogether) to understand why the paper's findings are important

Structure is usually:
    
1. Accepted state of knowledge in the field
2. Focus on a particular aspect of the field; often the sets of data that led directly to the work of this paper
3. Hypothes(i/e)s being tested
4. Conclusions (scientists don't really like surprise endings!)

**How to approach the introduction**:

> Grab a blank piece of paper

1. Take notes
2. Draw mini figures
3. Define vocabulary

> Answer these questions:

1. What is the accepted state of knowledge?
2. What data led directly to the work contained in the paper?
3. What is the hypothesis being tested?
4. What are hte basic conclusions?

### 5.4. Materials and Methods

These sections should be detailed enough for another scientist to replicate the work (volumes, times, which company material was used, purchased from where, etc.)

**Should you read the materials and methods?**

You can often skim over them before the results for a first reading. However, when you get to the results, you will need to flip back to them often to clarify how experiment was done.

### 5.5. Results

While the introduction poses the questions being asked, the results describe the outcome(s) of the experiments that were done to answer the questions.

Results are often simply stated with interpretations being reserved to the Discussion part.

Figures and tables allow the reader to see the outcomes of the experiments for themselves!

**How to read the results/figures**:

Read the text throughout. When a figure is referred to, examine the figure. Take notes, giving yourself a place to refer to about each figure. With each experiment/figure you should be able to explain:

1. basic procedure
2. question it sought to answer
3. results
4. conclusion
5. criticisms

**Legends** are mandatory to provide contextual information to figures and tables.

### 5.6. Discussion

Data is analyzed to show what the authors believe the data provides as evidence to their 'claims' (You don't have to agree with their interpretations!)

Findings are related to other findings in the field (contribution to knowledge, correction of errors, etc.) How is this work/result significant?

**Threats to validity**:

Discussion should list the internal and external threats to validity (e.g. influencing variables that have not been taken into account).

How to read a discussion:

**Take notes and answer these questions**:

1. What conclusions does the author draw? Be sure to separate fact from their opinion/interpretation?
2. Describe for yourself why these data are significant. Do they contribute to knowledge or correct errors?

### 5.7. References

Those are the papers cited in the text.

**What parts of the paper cite other papers?**

1. Introduction 
2. Materials and Methods
3. Discussions

The article, named 'Acceptance and Acceptability Criteria: A Literature Review' is a 2017 paper, published in January 2018, in the review ``Cognition, Technology, and Work`` (Springer, publied on the doi.org platform) in the trans-disciplinary field of psycho-ergonomics. Psycho-ergonomics draw from psychology, ergonomics, cognitive science and sociology to answer question related to the the modes pertaining to human-interface relationship. 

The paper itself is a literature review of the historic positions taken by field experts on two key ergonomics concepts: acceptance and acceptability. The goal of the study was to identify key approaches and factors in the study of those concepts, and draw from them outlines and common threads on which to proceed forward. 

The paper claims that 5 main classes of criteria (in total 142 criteria have been identified and condensed into 5 metacriteria/categories/classes) regarding the study of acceptance and acceptability exist within the corpus of knowledge of psycho-ergonomics, drawing from several historical approaches that have either influenced or fed each other across the evolution of the knowledge base of the field. 

The second claim of the paper is that 2 out of the five main metacriteria can be discarded as, based on the paper's weighted analysis method, the three remaining (which are solely related to the tool and tool experience) have a strong explanatory power.

Despite calling for a new reading grid for the field, with this literature review as both a sift and funnel, he paper limits its own reach by explaining that no single metacriteria can explain both acceptance and acceptability.
